```python
import os
# Aggregated Transactions
path="C:/Users/ABINASH/Desktop/ABINASH DS/VS Code/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
```

```python
import pandas as pd
import json

lis_t = {'State':[],'Year':[],'Quater':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}

for state in Agg_state_list:
  p_i = path + state + "/"
  Agg_y = os.listdir(p_i)
  for year in Agg_y:
    p_y = p_i + year + "/"
    Agg_q = os.listdir(p_y)
    for quater in Agg_q:
      p_q = p_y + quater
      Data = open(p_q,'r')
      D=json.load(Data)
      for s in D["data"]["transactionData"]:
        Name = s["name"]
        Type = s["paymentInstruments"][0]["type"]
        Count = s["paymentInstruments"][0]["count"]
        Amount = s["paymentInstruments"][0]["amount"]
        lis_t['Transaction_type'].append(Type)
        lis_t['Transaction_count'].append(Count)
        lis_t['Transaction_amount'].append(Amount)
        lis_t['State'].append(state)
        lis_t['Year'].append(year)
        lis_t['Quater'].append(int(quater.strip(".json")))

Agg_Data = pd.DataFrame(lis_t)


```
